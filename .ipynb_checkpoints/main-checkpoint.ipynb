#### 1. Importation des Bibliothèques Nécessaires

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

#### 2. Chargement des Données


```python
# Chargement du dataset local
file_path = "life_expectancy_data_averaged.csv"  
data = pd.read_csv(file_path, encoding='ISO-8859-1', sep = ",")
```

#### 3. Prétraitement des Données


- Renommer les colonnes en français

```python
data = data.rename(columns={
    'Country': 'Pays',
    'Region': 'Région',
    'Year': 'Année',
    'Infant_deaths': 'Décès_infantiles',
    'Under_five_deaths': 'Décès_sous_cinq_ans',
    'Adult_mortality': 'Mortalité_adulte',
    'Alcohol_consumption': 'Consommation_alcool',
    'Hepatitis_B': 'Hépatite_B',
    'Measles': 'Rougeole',
    'BMI': 'IMC',
    'Polio': 'Polio',
    'Diphtheria': 'Diphtérie',
    'Incidents_HIV': 'Incidents_HIV',
    'GDP_per_capita': 'PIB_par_habitant',
    'Population_mln': 'Population_en_millions',
    'Thinness_ten_nineteen_years': 'Maigreur_10_19_ans',
    'Thinness_five_nine_years': 'Maigreur_5_9_ans',
    'Schooling': 'Scolarité',
    'Economy_status': 'Statut_économique',
    'Life_expectancy': 'Espérance_de_vie'
})
```

- Mettre la colonne pays comme index

```python
data.set_index('Pays', inplace=True)
data.index.name = None
```

- Supprimer les colonnes inutiles

```python
data_clean = data.drop(["Région", "Année"], axis=1)
data_clean.head()
```

- Normaliser les données

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_clean)
data_scaled
```

#### 4. Analyse bivariées

- <span style="color:green; font-size:1em; font-weight:bold;">Calcul et Affichage de la Matrice de Corrélation</span>

```python
corr_matrix = data_clean.corr()
corr_matrix
```

- <span style="color:green; font-size:1em; font-weight:bold;"> Visualisation de la Heatmap des Corrélations</span>


```python
plt.figure(figsize=(11, 10))  
sns.heatmap(corr_matrix, 
            annot=True,         # Affiche les valeurs dans chaque case
            cmap="YlGnBu",      # Palette de couleurs
            square=True,        # Maintient les cellules carrées
            linewidths=0.5,     # Ajoute un contour entre les carrés
            cbar_kws={"shrink": 0.8})  # Réduit la taille de la barre des couleurs

plt.title('Heatmap des Corrélations')  
plt.show()
```

#### 5. Analyse multivariée avec l'ACP

- <span style="color:Green; font-size:1em; font-weight:bold;">Construction de l'ACP</span>

```python
pca = PCA()
X_pca = pca.fit_transform(data_scaled)

#nombre de composantes calculées
#print(pca.n_components_) #
X_pca
```

```python
# Supposons que 'data_scaled' soit vos données normalisées (centrées et réduites)
# Étape 1 : Calculer la matrice de covariance
cov_matrix = np.cov(data_scaled.T)

# Étape 2 : Calculer les valeurs propres et vecteurs propres
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Afficher les résultats
print("Valeurs propres :\n", eigenvalues)
print("Vecteurs propres :\n", eigenvectors)
```

```python
# Valeurs propres
eigenvalues = pca.explained_variance_
print("Valeurs propres :", eigenvalues)
eigenvalues

# Vecteurs propres
eigenvectors = pca.components_
print("Vecteurs propres :\n", eigenvectors)
```

- <span style="color:Green; font-size:1em; font-weight:bold;">Analyse des valeurs propres</span>

```python
n_dim = len(pca.explained_variance_)
comp = pd.DataFrame(
    {
        "Dimension" : ["Dim" + str(x + 1) for x in range(n_dim)], 
        "Valeur propre" : pca.explained_variance_,
        "% variance expliquée" : np.round(pca.explained_variance_ratio_ * 100),
        "% cum. var. expliquée" : np.round(np.cumsum(pca.explained_variance_ratio_) * 100)
    },
    columns = ["Dimension", "Valeur propre", "% variance expliquée", "% cum. var. expliquée"]
)
comp
```

- <span style="color:Green; font-size:1; font-weight:bold;">Scree plot pour choisir le nombre de composantes principales</span>

```python
g_comp = sns.barplot(x = "Dimension", 
                y = "% variance expliquée",
                data = comp)
g_comp.set(ylabel = "Variance expliquée (%)")
g_comp.figure.suptitle("Variance expliquée par dimension")
plt.axhline(y = 25, linewidth = .5, color = "dimgray", linestyle = "--") 
plt.text(3.25, 26, "25%")
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(range(0, 70, 10))

plt.show()
```

- <span style="color:Green; font-size:1em; font-weight:bold;"> Cercle des corrélations</span>

```python
#le champ components_ de l'objet ACP
print(pca.components_)
```

```python
n = len(pca.explained_variance_)
#valeur corrigée
eigval = (n-1)/n*pca.explained_variance_
print(eigval)
pca.explained_variance_
```

```python
#racine carrée des valeurs propres
sqrt_eigval = np.sqrt(eigval)
```

```python
#corrélation des variables avec les axes
corvar = np.zeros((n,n))
for k in range(n):
 corvar[:,k] = pca.components_[k,:] * sqrt_eigval[k]

#afficher la matrice des corrélations variables x facteurs
print(corvar)
```

```python
#on affiche pour les deux premiers axes
print(pd.DataFrame({'id':data_clean.columns,'COR_1':corvar[:,0],'COR_2':corvar[:,1]}))
```

```python
# Cercle des corrélations
fig, axes = plt.subplots(figsize=(8, 8))
axes.set_xlim(-1, 1)
axes.set_ylim(-1, 1)

# Affichage des étiquettes et des flèches associées aux variables
for j in range(n):
    # Coordonnées de la variable
    x, y = corvar[j, 0], corvar[j, 1]
    
    # Dessiner la flèche
    plt.arrow(0, 0, x, y, head_width=0.02, head_length=0.04, fc='red', ec='red', length_includes_head=True)
    
    # Afficher l'étiquette
    plt.annotate(data_clean.columns[j], (x, y), textcoords="offset points", xytext=(5, 5), ha='center')

# Ajouter les axes
plt.plot([-1, 1], [0, 0], color='silver', linestyle='-', linewidth=1)
plt.plot([0, 0], [-1, 1], color='silver', linestyle='-', linewidth=1)

# Ajouter un cercle
cercle = plt.Circle((0, 0), 1, color='blue', fill=False)
axes.add_artist(cercle)

# Affichage
plt.title('Cercle des corrélations')
plt.xlabel('Axe 1')
plt.ylabel('Axe 2')
plt.grid(False)
plt.show()
```

- <span style="color:Green; font-size:1em; font-weight:bold;"> La qualité de
représentation des individus sur les axes – cos2</span>

```python
#contribution des individus dans l'inertie totale
di = np.sum(data_scaled**2,axis=1)
print(pd.DataFrame({'ID':data_clean.index,'d_i':di}))
```

```python
#qualité de représentation des individus - COS2
cos2 = X_pca**2
for j in range(n):
 cos2[:,j] = cos2[:,j]/di
print(pd.DataFrame({'id':data_clean.index,'COS2_1':cos2[:,0],'COS2_2':cos2[:,1]}))
```

```python
# Extraction des dimensions et des contributions
dimension_1 = X_pca[:, 0]  # Première composante principale
dimension_2 = X_pca[:, 1]  # Deuxième composante principale
pays = data_clean.index  # Noms des individus (villes)
cos2_2 = cos2[:, 1]        # Contribution à la première composante pour la coloration

# Création du graphique avec une palette adaptée
plt.figure(figsize=(9, 5))
scatter = plt.scatter(dimension_1, dimension_2, c=cos2_2, cmap="coolwarm", s=20)
colors = scatter.to_rgba(cos2_2)  # Couleurs utilisées pour les points

# Ajouter les noms des pays au-dessus des points avec la même couleur
for i, p in enumerate(pays):
    plt.text(
        dimension_1[i], 
        dimension_2[i] + 0.2, 
        p, 
        fontsize=6, 
        ha="center", 
        color=colors[i]
    )

# Ajouter les axes avec des lignes pointillées
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")

# Ajouter les labels et le titre
plt.xlabel(f"Dim 1 ({pca.explained_variance_ratio_[0] * 100:.2f}%)", fontsize=12)
plt.ylabel(f"Dim 2 ({pca.explained_variance_ratio_[1] * 100:.2f}%)", fontsize=12)
plt.title("Graphe des individus de l'ACP – habillage selon (COS2_2)", fontsize=12, color="blue")

# Ajuster les ticks sur l'axe des ordonnées
y_ticks = np.arange(int(min(dimension_2)) - 2, int(max(dimension_2)) + 3, 2)
plt.yticks(y_ticks)

# Ajouter une barre de couleur pour les contributions
cbar = plt.colorbar(scatter)
cbar.set_label("COS2_2", fontsize=12)

# Afficher le graphique
plt.tight_layout()
plt.show()
```

- <span style="color:Green; font-size:1em; font-weight:bold;"> Contribution des individus aux axes (CTR)</span>

```python
#contributions aux axes
ctr = X_pca**2
for j in range(n):
 ctr[:,j] = ctr[:,j]/(n*eigval[j])

print(pd.DataFrame({'id':data_clean.index,'CTR_1':ctr[:,0],'CTR_2':ctr[:,1]}))
```

```python
# Extraction des dimensions et des contributions
dimension_1 = X_pca[:, 0]  # Première composante principale
dimension_2 = X_pca[:, 1]  # Deuxième composante principale
pays = data_clean.index  # Noms des individus (villes)
ctr_1 = ctr[:, 0]  # Contribution à la première composante pour la coloration

# Création du graphique avec une palette adaptée
plt.figure(figsize=(9, 5))
scatter = plt.scatter(dimension_1, dimension_2, c=ctr_1, cmap="coolwarm", s=30)
colors = scatter.to_rgba(ctr_1)  # Couleurs utilisées pour les points

# Ajouter les noms des pays au-dessus des points avec la même couleur
for i, p in enumerate(pays):
    plt.text(
        dimension_1[i], 
        dimension_2[i] + 0.2, 
        p, 
        fontsize=6, 
        ha="center", 
        color=colors[i]
    )

# Ajouter les axes avec des lignes pointillées
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")

# Ajouter les labels et le titre
plt.xlabel(f"Dim 1 ({pca.explained_variance_ratio_[0] * 100:.2f}%)", fontsize=12)
plt.ylabel(f"Dim 2 ({pca.explained_variance_ratio_[1] * 100:.2f}%)", fontsize=12)
plt.title("Graphe des individus de l'ACP – habillage selon la contribution (CTR_1)", fontsize=12, color="blue")

# Ajuster les ticks sur l'axe des ordonnées
y_ticks = np.arange(int(min(dimension_2)) - 2, int(max(dimension_2)) + 3, 2)
plt.yticks(y_ticks)

# Ajouter une barre de couleur pour les contributions
cbar = plt.colorbar(scatter)
cbar.set_label("Contribution (CTR_1)", fontsize=12)

# Afficher le graphique
plt.tight_layout()
plt.show()
```
